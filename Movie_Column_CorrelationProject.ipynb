```python
# Import Libraries

import pandas as pd
import seaborn as sns
import numpy as np

import matplotlib
import matplotlib.pyplot as plt
plt.style.use('ggplot')
from matplotlib.pyplot import figure

%matplotlib inline
matplotlib.rcParams['figure.figsize'] = (12,8) #Adjust the configuration of the plots we will create

#Read in the data

df= pd.read_csv(r'C:\Users\gn799\Videos\Kaggle\movies.csv')
```

```python
# Let's look at the data

df.head()
```

```python
# Lets see if there is some missing data

for col in df.columns:
    pct_missing = np.mean(df[col].isnull())
    print('{} - {}%'.format(col,pct_missing))
```

```python
# Data types for our collumns

df.dtypes
```

```python
# Change data type of columns
df['budget'].dtype
```

```python
df['budget'] = df['budget'].astype('Int64')
df['gross'] = df['gross'].astype('Int64')
```

```python
df.dtypes
```

```python
df.head()
```

```python
#Create correct Year column

#Convert released collumn to string
df['released'] = df['released'].astype('str')

# Create final_Year collumn and extract year only
df['final_year'] = df['released'].str.extract(r'(\d{4})')

#delete collumns
df = df.drop(columns=['year_correct', 'cleaned_date','correct_year'])

df
```

```python
df.dtypes
```

```python
df.sort_values(by=['gross'],inplace = False, ascending = False)
```

```python
#arrange data set in ascending order according to Gross

df= df.sort_values(by=['gross'],inplace = False, ascending = False)
```

```python
# Drop any duplicated

df['company'].drop_duplicates().sort_values(ascending= False)
```

```python
correlation_matrix = df.corr (method = 'kendall', numeric_only=True)
sns.heatmap(correlation_matrix, annot = True)

plt.title('Correlation Matric for Numeric Features')

plt.ylabel('Movie Features')
plt.xlabel('Movie Features')

plt.show()
```

```python
#Look at Company

df.head()
```

```python
df_numerized = df

for col_name in df_numerized.columns:
    if (df_numerized[col_name].dtype == 'object'):
        df_numerized[col_name] = df_numerized[col_name].astype('category')
        df_numerized[col_name] = df_numerized[col_name].cat.codes

df_numerized
```

```python
correlation_matrix = df_numerized.corr (method = 'kendall', numeric_only=True)
sns.heatmap(correlation_matrix, annot = True)

plt.title('Correlation Matric for Numeric Features')

plt.ylabel('Movie Features')
plt.xlabel('Movie Features')

plt.show()
```

```python
df_numerized.corr()
```

```python
correlation_mat = df_numerized.corr()
corr_pairs = correlation_mat.unstack()

corr_pairs
```

```python
sorted_pairs = corr_pairs.sort_values()
sorted_pairs
```

```python
high_corr = sorted_pairs[(sorted_pairs) > 0.4]
high_corr
```
